```python
import matplotlib.pyplot as plt

# Remplacez 'Suivi_Actualite_Sportive' par le nom réel de la colonne
colonne_actualite_sportive = 'Suivez-vous l\'actualité sportive ?'

# Calculer la répartition des réponses
repartition = df[colonne_actualite_sportive].value_counts()

# Créer le diagramme circulaire
plt.figure(figsize=(8, 8))
plt.pie(repartition, labels=repartition.index, autopct='%1.1f%%', startangle=140)
plt.title('Répartition des Réponses à la Question "Suivez-vous l\'actualité sportive ?"')

# Ajouter une légende
plt.legend(repartition.index, title="Réponses", loc="best")

plt.show()
```

## Analyse du Suivi de l'Actualité Sportive

### Interprétation du Diagramme
Le diagramme circulaire représente les réponses à la question "Suivez-vous l'actualité sportive ?". Sur l'ensemble des répondants, une grande majorité, soit 81,4%, ont indiqué qu'ils suivent l'actualité sportive. En revanche, 18,6% des répondants ne suivent pas l'actualité sportive.

### Conclusion
Cette répartition suggère que l'intérêt pour le sport et l'actualité sportive est assez élevé parmi la population sondée. L'actualité sportive peut donc être considérée comme un élément important de la vie quotidienne pour la plupart des gens dans cet échantillon. Les organisations sportives, les annonceurs et les médias pourraient utiliser cette information pour cibler et engager la majorité des personnes qui montrent un intérêt pour les sports. Cependant, il reste près d'un cinquième de l'échantillon qui ne suit pas l'actualité sportive, ce qui pourrait représenter une opportunité pour développer des stratégies afin de capter leur intérêt ou de comprendre les raisons de cette désaffection.


```python
# Remplacez 'colonne_frequence' par le nom réel de la colonne dans votre fichier Excel
colonne_frequence = 'A quelle fréquence pratiquez-vous une activité physique ou sportive ?'

# Nettoyage des données si nécessaire
# Par exemple: df[colonne_frequence].fillna('Non spécifié', inplace=True)

# Création du graphique
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=colonne_frequence)
plt.title("Fréquence de pratique d'activité physique ou sportive")
plt.xticks(rotation=45)
plt.show()
```

## Analyse de la Fréquence de la Pratique d'Activité Physique ou Sportive

### Interprétation du Diagramme
Le diagramme en barres illustre la fréquence à laquelle les individus pratiquent une activité physique ou sportive. Il est évident que la pratique une fois par semaine est la plus courante parmi les répondants, suivie de près par ceux qui pratiquent trois fois par semaine. Les individus qui s'engagent dans une activité physique ou sportive deux fois par semaine représentent le troisième groupe le plus important. En revanche, ceux qui pratiquent plus de quatre fois par semaine ou quatre fois par semaine sont moins nombreux.

### Conclusion
Cette distribution suggère que la plupart des personnes interrogées ont intégré une certaine forme d'activité physique ou sportive dans leur routine hebdomadaire, avec une prédominance significative pour la pratique hebdomadaire. Cela peut refléter des contraintes de temps ou des préférences pour une fréquence modérée d'exercice, ce qui peut être suffisant pour maintenir un niveau de forme physique de base. Toutefois, une minorité plus petite se livre à des activités sportives avec une fréquence beaucoup plus élevée, ce qui pourrait indiquer un engagement plus sérieux envers la forme physique ou des objectifs sportifs spécifiques. Les décideurs politiques et les planificateurs urbains pourraient utiliser ces informations pour s'assurer que les installations sportives répondent à la demande des citoyens et pour promouvoir des programmes visant à augmenter la fréquence de l'activité physique pour améliorer la santé publique.


```python
# Remplacez cette partie par votre code de chargement de données
data = {
    'Accès_Installations': ['Oui', 'Non', 'Oui', 'Non', 'Oui', 'Oui', 'Non', 'Oui', 'Oui', 'Non'],
    'Lieu_Residence': ['Ville', 'Campagne', 'Ville', 'Ville', 'Campagne', 'Ville', 'Campagne', 'Ville', 'Campagne', 'Ville']
}
df = pd.DataFrame(data)

# Création du graphique avec une différenciation par lieu de résidence
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Accès_Installations', hue='Lieu_Residence')
plt.title("Accès à des installations sportives dans la localité par lieu de résidence")
plt.xlabel("Accès aux installations")
plt.ylabel("Nombre de personnes")
plt.show()

```

## Analyse de l'Accès aux Installations Sportives par Lieu de Résidence

### Interprétation du Diagramme
Le diagramme en barres compare l'accès aux installations sportives en fonction du lieu de résidence, distinguant les réponses des habitants de la ville de celles des habitants de la campagne. Il est évident que les personnes vivant en ville rapportent un accès plus élevé aux installations sportives que celles vivant à la campagne. Cela est indiqué par la barre plus élevée pour les réponses "Oui" des résidents de la ville. En revanche, pour les réponses "Non", il y a une répartition presque égale entre les résidents de la ville et ceux de la campagne, avec une légère prédominance pour la campagne.

### Conclusion
Ces données suggèrent que les installations sportives sont perçues comme étant plus accessibles en milieu urbain qu'en milieu rural. Cela peut refléter une concentration plus dense des installations sportives dans les zones urbaines, qui sont généralement mieux desservies en termes d'infrastructures. Les autorités locales et les planificateurs peuvent considérer ces informations comme un indicateur du besoin d'améliorer l'accès aux installations sportives en milieu rural ou de développer des programmes qui favorisent l'activité physique en dehors des installations traditionnelles, notamment pour les résidents de la campagne.


```python
# Remplacez 'Duree_Acces' par le nom réel de la colonne
colonne_temps_acces = 'Combien de temps mettez-vous pour accéder à ces installations ?'

plt.figure(figsize=(10, 6))
sns.histplot(df[colonne_temps_acces], kde=False)  # Retirez kde=True si vous ne voulez pas de la courbe de densité
plt.title('Distribution du Temps d\'Accès aux Installations Sportives')
plt.xlabel('Temps d\'Accès (minutes)')
plt.ylabel('Nombre de Personnes')
plt.show()
```

## Analyse du Temps d'Accès aux Installations Sportives

### Interprétation du Diagramme
Le diagramme en barres représente la distribution du temps nécessaire aux personnes pour accéder aux installations sportives. La majorité écrasante des répondants indiquent un temps d'accès de moins de 15 minutes, ce qui suggère que les installations sportives sont relativement proches de leur résidence ou lieu de travail. La deuxième tranche la plus commune est celle des 15 à 30 minutes, suivie par le groupe des 30 minutes à 1 heure. Notablement moins nombreux sont ceux qui mettent entre 1 et 2 heures, et un nombre encore plus réduit de répondants indique un temps d'accès de plus de 2 heures.

### Conclusion
Les données révèlent que les installations sportives sont généralement bien situées pour assurer un accès rapide pour la plupart des personnes. Cela peut être indicatif d'une bonne planification urbaine et de la présence suffisante d'installations sportives dans les zones à forte densité de population. Néanmoins, il existe un petit segment de la population pour lequel l'accès est significativement plus long, dépassant 1 heure. Ces informations pourraient être utilisées par les autorités locales pour identifier les zones sous-desservies et envisager le développement de nouvelles installations ou l'amélioration des réseaux de transport pour faciliter un accès plus rapide et équitable à l'infrastructure sportive.

## Analyse des Opinions sur les Politiques d'Aménagement Sportives du Sénégal

Le graphique présente les opinions sur les politiques d'aménagement sportives du Sénégal depuis l'an 2000 en fonction de la fréquence à laquelle les individus pratiquent une activité sportive. Les personnes qui pratiquent une fois par semaine semblent avoir l'opinion la plus positive, avec le nombre le plus élevé d'avis favorables. Ceux qui s'entraînent trois fois par semaine suivent de près. Intéressant, les individus qui s'exercent deux fois ou plus de quatre fois par semaine ont exprimé moins d'opinions, tandis que ceux qui s'entraînent quatre fois par semaine ont le nombre le plus faible d'opinions exprimées, ce qui pourrait indiquer un niveau d'indifférence ou un manque d'information suffisante pour formuler une opinion.

### Conclusion
Cette distribution peut suggérer que les personnes les plus actives physiquement, en particulier celles qui pratiquent une activité une fois par semaine, sont soit les plus satisfaites des politiques d'aménagement sportives, soit les plus conscientes de leur existence et de leur impact. Il est possible que ceux qui s'exercent plus fréquemment aient des attentes différentes ou plus élevées concernant ces politiques, ce qui pourrait expliquer pourquoi leurs opinions sont moins nombreuses ou moins positives. L'implication pour les décideurs politiques est que l'augmentation de la fréquence des activités sportives ne se traduit pas nécessairement par une perception plus positive des politiques sportives, soulignant la nécessité d'une approche ciblée pour répondre aux besoins des sportifs réguliers et très actifs.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assurez-vous que ces noms correspondent exactement aux noms de colonnes de votre DataFrame
colonne_accessibilite = 'Pensez-vous  que les installations sportives sont accessibles pour les personnes à mobilité réduite ?'
colonne_tranche_age = 'Quelle tranche d\'âge appartenez-vous ? '

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=colonne_accessibilite, hue=colonne_tranche_age, order=df[colonne_accessibilite].value_counts().index)
plt.title('Accessibilité des Installations Sportives par Tranche d\'Âge')
plt.xlabel('Accessibilité pour les Personnes à Mobilité Réduite')
plt.ylabel('Nombre de Réponses')
plt.legend(title='Tranche d\'Âge')
plt.show()
```

## Analyse de l'Accessibilité des Installations Sportives pour les Personnes à Mobilité Réduite

### Interprétation du Diagramme
Le diagramme en barres illustre les perceptions de l'accessibilité des installations sportives pour les personnes à mobilité réduite, ventilées par tranches d'âge. Il est remarquable que la grande majorité des répondants de la tranche d'âge 30-60 ans perçoivent ces installations comme étant inaccessibles. De manière surprenante, les autres tranches d'âge, y compris les jeunes de moins de 15 ans, les 15-30 ans, et les seniors de plus de 60 ans, semblent considérer l'accessibilité de façon beaucoup plus positive, bien que le nombre de réponses pour ces groupes soit nettement inférieur.

### Conclusion
Ces résultats suggèrent qu'il peut y avoir une sensibilisation ou une expérience significativement différente de l'accessibilité des installations sportives entre les générations. La tranche d'âge 30-60 ans, qui est susceptible d'inclure des parents de jeunes enfants et des individus actifs professionnellement, peut être plus critique ou avoir des attentes plus élevées en matière d'accessibilité. Les données indiquent un besoin potentiel de se concentrer sur l'amélioration de l'accessibilité pour cette tranche d'âge spécifique. En outre, étant donné le nombre relativement faible de réponses des autres tranches d'âge, il pourrait être utile d'approfondir l'enquête pour comprendre pleinement leurs perspectives et expériences. Cette information pourrait être cruciale pour les organismes de planification sportive et les décideurs pour cibler les améliorations d'accessibilité là où elles sont le plus nécessaires.
